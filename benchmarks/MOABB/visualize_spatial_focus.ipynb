```python
from train import load_hparams_and_dataset_iterators
import torch

import matplotlib.pyplot as plt
from mne.viz import plot_topomap
import mne.channels
import glob
import numpy as np
```

```python
subject, session = 0, 0
weights_array = []
for subject in range(9):
        hparams, ds = load_hparams_and_dataset_iterators('results/MotorImagery/BNCI2014004/SpatialEEGNet_cross/hopt/best_hparams.yaml', {}, {
            "data_folder": "/home/ubuntu/mne_data",
            "output_folder": "/tmp/foo",
            "cached_data_folder": "/home/ubuntu/mne_data/pkl",
            "target_subject_idx": subject,
            "target_session_idx": 1,
            "data_iterator_name": "leave-one-subject-out"
        })
        
        sample = next(iter(ds['train'].dataset))
        
        ckpts = glob.glob(f'results/MotorImagery/BNCI2014004/SpatialEEGNet_cross/hopt/best/BJUZME/*/*/leave-one-subject-out/sub-{str(subject+1).zfill(3)}/save/*/model.ckpt')
        for ckpt in ckpts:
            hparams['model'].load_state_dict(torch.load(ckpt, map_location=torch.device('cpu')))
            ch_positions = sample.pos
            ch_positions_orig = sample.pos
            ch_positions = (
                2 * ((ch_positions - ch_positions.min(0).values) / (ch_positions.max(0).values - ch_positions.min(0).values) - 0.5)
            )
            weights = hparams['model'].spatial_focus.similarity_module(ch_positions.float()).div_(hparams["spatial_focus_tau"]).softmax(-2)
            weights_array.append(weights)
```

```python
weights = torch.hstack(weights_array).T.cpu().detach().numpy()
weights
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=hparams['projection_dim'])
km.fit(np.log(weights))
```

```python
rows = int(np.ceil(hparams['projection_dim'] / 6))
fig, axes = plt.subplots(rows, 6, figsize=(2*6, 2*rows), layout="constrained")
for ax in axes.flat:
    ax.axis('off')
for i, ax in zip(range(km.cluster_centers_.shape[0]), axes.flat):
    plot_topomap(data=np.exp(km.cluster_centers_[i]), pos=ch_positions_orig[:, [0, 1]], axes=ax, show=False)

plt.savefig("bnci2014_004_clusters.pdf", format="PDF")
print(km.inertia_ / len(weights))
```
